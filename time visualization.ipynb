```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# from mpl_toolkits.basemap import Basemap
# from matplotlib import cm
%matplotlib inline
import datetime
import numpy as np
```

```python
uber_2014 = pd.read_csv(r'C:\Users\zhenz\Downloads\uber_trips_2014.csv\uber_trips_2014.csv', parse_dates=['pickup_datetime'])
# uber_2015 = pd.read_csv(r'C:/Users/zhenz/Downloads/uber_trips_2015.csv')
```

```python
df_H = uber_2014.pickup_datetime.value_counts().resample('H', how='sum')
df_D = uber_2014.pickup_datetime.value_counts().resample('D', how='sum')
```

```python
filter_2014 = np.logical_and(pd.to_datetime(uber_2014.pickup_datetime) < datetime.date(2014,5,12),
                             pd.to_datetime(uber_2014.pickup_datetime) >= datetime.date(2014,5,5))
```

```python
week = uber_2014.loc[filter_2014,:].pickup_datetime.value_counts().resample('H', how='sum')
```

```python
week[week.index.hour.isin(range(5, 10))].resample('D', how='sum').plot()
week[week.index.hour.isin(range(10, 16))].resample('D', how='sum').plot()
week[week.index.hour.isin(range(16, 19))].resample('D', how='sum').plot()
week[week.index.hour.isin(range(19, 24))].resample('D', how='sum').plot()

plt.legend(['Morning Rush Hour', 'Noon', 'Evening Rush Hour', 'Night'])
```

```python
filter_mon = np.logical_and(pd.to_datetime(uber_2014.pickup_datetime) < datetime.date(2014,5,6),
                               pd.to_datetime(uber_2014.pickup_datetime) >= datetime.date(2014,5,5))
filter_tue = np.logical_and(pd.to_datetime(uber_2014.pickup_datetime) < datetime.date(2014,5,7),
                             pd.to_datetime(uber_2014.pickup_datetime) >= datetime.date(2014,5,6))
filter_wed = np.logical_and(pd.to_datetime(uber_2014.pickup_datetime) < datetime.date(2014,5,8),
                               pd.to_datetime(uber_2014.pickup_datetime) >= datetime.date(2014,5,7))
filter_thu = np.logical_and(pd.to_datetime(uber_2014.pickup_datetime) < datetime.date(2014,5,9),
                               pd.to_datetime(uber_2014.pickup_datetime) >= datetime.date(2014,5,8))
filter_fri = np.logical_and(pd.to_datetime(uber_2014.pickup_datetime) < datetime.date(2014,5,10),
                               pd.to_datetime(uber_2014.pickup_datetime) >= datetime.date(2014,5,9))
filter_sat = np.logical_and(pd.to_datetime(uber_2014.pickup_datetime) < datetime.date(2014,5,11),
                               pd.to_datetime(uber_2014.pickup_datetime) >= datetime.date(2014,5,10))
filter_sun = np.logical_and(pd.to_datetime(uber_2014.pickup_datetime) < datetime.date(2014,5,12),
                               pd.to_datetime(uber_2014.pickup_datetime) >= datetime.date(2014,5,11))
```

```python
plt.plot(uber_2014.loc[filter_mon, :].pickup_datetime.value_counts().resample('H', how='sum').values)
plt.plot(uber_2014.loc[filter_tue, :].pickup_datetime.value_counts().resample('H', how='sum').values)
plt.plot(uber_2014.loc[filter_wed, :].pickup_datetime.value_counts().resample('H', how='sum').values)
plt.plot(uber_2014.loc[filter_thu, :].pickup_datetime.value_counts().resample('H', how='sum').values)
plt.plot(uber_2014.loc[filter_fri, :].pickup_datetime.value_counts().resample('H', how='sum').values)
plt.plot(uber_2014.loc[filter_sat, :].pickup_datetime.value_counts().resample('H', how='sum').values)
plt.plot(uber_2014.loc[filter_sun, :].pickup_datetime.value_counts().resample('H', how='sum').values)
plt.legend(['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
```
